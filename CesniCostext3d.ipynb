```python
def find_nota(table, nota):
    nota_row = -1
    nota_column = -1
    for row in range(table.shape[0]):
        for column in range(table.shape[1]):
            if table.iat[row,column] == nota:
                nota_row = row
                nota_column = column
                       
    return nota_row, nota_column   
```

```python
def path_finder_f(nota_1, nota_2, p_up_weight,p_down_weight,b_weight, d_weight, g_weight, k_weight):
    
    (table_1, row_1, column_1) = find_nota_in_table(nota_1)
    #print("Nota", nota_1,"Table", table_1,"Row", row_1,"Column",column_1)
    #print("Main-X", arr_main[table_1-1][0], "Main-Y", arr_main[table_1-1][1])    
    #print("X kayna", row_1-arr_main[table_1-1][0], "Y kayna", column_1-arr_main[table_1-1][1])
    
    
    (table_2, row_2, column_2) = find_nota_in_table(nota_2)
    #print("Nota", nota_2,"Table", table_2,"Row", row_2,"Column",column_2)
    #print("Main-X", arr_main[table_2-1][0], "Main-Y", arr_main[table_2-1][1])  
    #print("X kayna", row_2-arr_main[table_2-1][0], "Y kayna", column_2-arr_main[table_2-1][1])
    
    delta_south_center = (row_2-arr_main[table_2-1][0]) - (row_1-arr_main[table_1-1][0]) 
    delta_east_center = (column_2-arr_main[table_2-1][1]) - (column_1-arr_main[table_1-1][1])
    
    #print("delta_guney_center", delta_south_center)
    #print("delta_dogu_center", delta_east_center)
       
    
    
    delta_table = table_2 - table_1
    delta_column = column_2 - column_1
    delta_row = row_2 - row_1
  
    delta_T1_T2 = 1
    delta_T2_T3 = 2
    delta_T3_T4 = 1
    delta_T4_T5 = 2
    
    
    p_up = 0
    p_down = 0
    
    table_direction_changed = False
    if (table_1==table_2):
        p_direction = "Panel is the same"
        p_weight = 0
    elif(table_1<table_2):
        p_direction = "Panel Goes Down"
        p_weight = p_down_weight
    else:
        p_direction = "Panel Goes Up"
        p_weight = p_up_weight
        table_direction_changed = True
        temp = table_1
        table_1 = table_2
        table_2 = temp
 
    
    delta_east = 0
    delta_south = 0
    if (table_1==table_2):
        p = 0
        delta_east = 0
        p_direction = "is the same"
        delta_south = 0
    elif (table_1==1) & (table_2==2):
        p = 1
        delta_south = 0
        delta_east = delta_T1_T2
    elif (table_1==1) & (table_2==3):
        p = 2
        delta_south = 0
        delta_east = delta_T1_T2 + delta_T2_T3
    elif (table_1==1) & (table_2==4):
        p = 3
        delta_south = 0
        delta_east = delta_T1_T2 + delta_T2_T3 + delta_T3_T4
    elif (table_1==1) & (table_2==5):
        p = 4
        delta_south = 0
        delta_east = delta_T1_T2 + delta_T2_T3 + delta_T3_T4 + delta_T4_T5
    elif (table_1==2) & (table_2==3):
        p = 1
        delta_south = 0
        delta_east = delta_T2_T3
    elif (table_1==2) & (table_2==4):
        p = 2
        delta_south = 0
        delta_east = delta_T2_T3 + delta_T3_T4
    elif (table_1==2) & (table_2==5):
        p = 3
        delta_south = 0
        delta_east = delta_T2_T3 + delta_T3_T4 + delta_T4_T5
    elif (table_1==3) & (table_2==4):
        p = 1
        delta_south = 0
        delta_east = delta_T3_T4
    elif (table_1==3) & (table_2==5):
        p = 2    
        delta_south = 0
        delta_east = delta_T3_T4 + delta_T4_T5
    elif (table_1==4) & (table_2==5):
        p = 1 
        delta_south = 0
        delta_east = delta_T4_T5    
   
    
    if(table_direction_changed == False):
        delta_d = delta_east + delta_east_center
        delta_g = delta_south + delta_south_center
        if(delta_d>0):
            d = delta_d
            b = 0
        elif(delta_d<0):
            b = -delta_d
            d = 0
        elif(delta_d==0):
            b = 0
            d = 0 
        if(delta_g>0):
            g = delta_g
            k = 0
        elif(delta_g<0):
            k = -delta_g
            g = 0
        elif(delta_g==0):
            k = 0
            g = 0     
            
            
    elif(table_direction_changed == True):
        delta_d = (delta_east - delta_east_center)
        delta_g = delta_south - delta_south_center
        if(delta_d>0):
            b = delta_d
            d= 0
        elif(delta_d<0):
            d = -delta_d
            b = 0
        elif(delta_d==0):
            b = 0
            d = 0
        if(delta_g>0):
            k = delta_g
            g = 0
        elif(delta_g<0):
            g = -delta_g
            k = 0
        elif(delta_g==0):
            k = 0
            g = 0     
    
    
    if (p_direction == "Panel is the same"):
        p_up = 0
        p_down = 0
    elif(p_direction == "Panel Goes Down"):
        p_down = p
        p_up = 0
    elif(p_direction == "Panel Goes Up"):
        p_down = 0
        p_up = p
    
    change_text = p_direction, "Up:", p_up ,"Down", p_down,  "East", d, "West", b, "South", g ,"North", k     
    
    
     
    cost = p_up_weight * p_up + p_down_weight * p_down + b_weight * b + d_weight * d + g_weight * g + k_weight * k
    
    
    return change_text, cost, p_up, p_down, b, d, g, k
    
 
```

```python
def cost_perde_cesni_p2p(starting_perde1, cesni1, starting_perde2, cesni2): 
   
    starting_perde = starting_perde1
    starting_perde_df = perdes_df[perdes_df["PerdeTo"]==starting_perde]
    perde_cesni_perdes_df = perdecesni_df[perdecesni_df["Perde"]==starting_perde]
    
    perde_cesni_df = perde_cesni_perdes_df[perde_cesni_perdes_df.Cesni==cesni1]
    perde_cesni_1 = perde_cesni_df.values[0].tolist()[2:7]
    perde_cesni_1 = [x for x in perde_cesni_1 if str(x) != 'nan']
    #print(perde_cesni_1)
    
    starting_perde = starting_perde2
    starting_perde_df = perdes_df[perdes_df["PerdeTo"]==starting_perde]
    perde_cesni_perdes_df = perdecesni_df[perdecesni_df["Perde"]==starting_perde]
    
    perde_cesni_df = perde_cesni_perdes_df[perde_cesni_perdes_df.Cesni==cesni2]
    perde_cesni_2 = perde_cesni_df.values[0].tolist()[2:7]
    perde_cesni_2 = [x for x in perde_cesni_2 if str(x) != 'nan']
    #print(perde_cesni_2)
    
    cesni_pc2pc_euclidean_cost = 0
    cesni_pc2pc_manhattan_cost = 0
 
    for i in range(min(len(perde_cesni_1), len(perde_cesni_2))):
        cesni_pc2pc_euclidean_cost += euclidean_distance_matrix_arr_df[perde_cesni_1[i]][perde_cesni_2[i]]
        cesni_pc2pc_manhattan_cost += manhattan_distance_matrix_arr_df[perde_cesni_1[i]][perde_cesni_2[i]]
        #print(perde_cesni_1[i], perde_cesni_2[i])
    
    if(len(perde_cesni_1)>len(perde_cesni_2)):
        for j in range(i+1,len(perde_cesni_1)):
            cesni_pc2pc_euclidean_cost += euclidean_distance_matrix_arr_df[start_perde][perde_cesni_1[j]]
            cesni_pc2pc_manhattan_cost += manhattan_distance_matrix_arr_df[start_perde][perde_cesni_1[j]]
    elif(len(perde_cesni_1)<len(perde_cesni_2)):
        for j in range(i+1,len(perde_cesni_2)):
            cesni_pc2pc_euclidean_cost += euclidean_distance_matrix_arr_df[start_perde][perde_cesni_2[j]]
            cesni_pc2pc_manhattan_cost += manhattan_distance_matrix_arr_df[start_perde][perde_cesni_2[j]]

    return cesni_pc2pc_euclidean_cost, cesni_pc2pc_manhattan_cost
```

```python
def cesni_move(perde_1, perde_2, cesni_1, cesni_2):

    perde_start = perdecesni_df[perdecesni_df.Perde==perde_1]
    perde_end   = perdecesni_df[perdecesni_df.Perde==perde_2]
    cesni_start = perde_start[perde_start.Cesni==cesni_1]
    cesni_end   = perde_end[perde_end.Cesni==cesni_2]

    #print("perde_start", perde_start)
    
    #print("perde_end", perde_end)
    #print("------------------")
    #print("cesni start", cesni_start, len(cesni_start))
    #print("cesni_end", cesni_end, len(cesni_end))
    #print("------------------")

    if(cesni_start.Perde4.isna().item() or cesni_end.Perde4.isna().item()):
        cesni_transform_length = 3 
    elif(cesni_start.Perde5.isna().item() or cesni_end.Perde5.isna().item()):
        cesni_transform_length = 4
    else:
        cesni_transform_length = 5


    #print("Length of Cesni Transform:",cesni_transform_length)
    
    change_text_c, cost_c, p_up_c, p_down_c, b_c, d_c, g_c, k_c = path_finder_c(perde_1, perde_2, p_up_weight, p_down_weight, b_weight, d_weight, g_weight, k_weight)
    change_text_f, cost_f, p_up_f, p_down_f, b_f, d_f, g_f, k_f = path_finder_f(perde_1, perde_2, p_up_weight, p_down_weight, b_weight, d_weight, g_weight, k_weight)
    
    if cost_c<cost_f:
        cost = cost_c
        change_text = change_text_c
        p_up = p_up_c
        p_down = p_down_c
        b = b_c
        d = d_c
        g = g_c
        k = k_c
    elif cost_c>cost_f:
        cost = cost_f
        change_text = change_text_f
        b = b_f
        d = d_f
        g = g_f
        k = k_f
    elif cost_c==cost_f:    
        cost = cost_f
        change_text = change_text_f
        b = b_f
        d = d_f
        g = g_f
        k = k_f
    
    cost_perde = 0
    cost_perde = cost
    #print("Cost of perde", cost_perde)

    cost_cesni = 0
    total_cost = 0
    
    
    for i in range(cesni_transform_length):
        
        if(cesni_start.iloc[0][2+i]!=cesni_end.iloc[0][2+i]):
            #print(f"Transform in Perde{i+1} | Start Perde:", cesni_start.iloc[0][2+i], "moves to End Perde:", cesni_end.iloc[0][2+i])
            #print(cesni_start.iloc[0][2+i])
            #print(cesni_end.iloc[0][2+i])
            
            change_text_c, cost_c, p_up_c, p_down_c, b_c, d_c, g_c, k_c = path_finder_c(cesni_start.iloc[0][2+i].strip(), cesni_end.iloc[0][2+i].strip(), p_up_weight, p_down_weight, b_weight, d_weight, g_weight, k_weight)
            change_text_f, cost_f, p_up_f, p_down_f, b_f, d_f, g_f, k_f = path_finder_f(cesni_start.iloc[0][2+i].strip(), cesni_end.iloc[0][2+i].strip(), p_up_weight, p_down_weight, b_weight, d_weight, g_weight, k_weight)
            if cost_c<cost_f:
                cost = cost_c
                change_text = change_text_c
                p_up = p_up_c
                p_down = p_down_c
                b = b_c
                d = d_c
                g = g_c
                k = k_c
            elif cost_c>cost_f:
                cost = cost_f
                change_text = change_text_f
                b = b_f
                d = d_f
                g = g_f
                k = k_f
            elif cost_c==cost_f:    
                cost = cost_f
                change_text = change_text_f
                b = b_f
                d = d_f
                g = g_f
                k = k_f

            #print(change_text)
            #print("Cost", cost)

            cost_cesni += cost

        #print("Çeşni Cost:", cost_cesni)    
        total_cost = cost_perde + cost_cesni
    
        #print("Total Cost:", total_cost)
        
        
        
    return total_cost

```
