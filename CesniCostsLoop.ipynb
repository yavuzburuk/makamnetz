```python
def find_nota(table, nota):
    nota_row = -1
    nota_column = -1
    for row in range(table.shape[0]):
        for column in range(table.shape[1]):
            if table.iat[row,column] == nota:
                nota_row = row
                nota_column = column
                       
    return nota_row, nota_column  
```

```python
def cost_matrixes(x_coord, y_coord, z_coord):

    w, h = len(x_coord), len(x_coord)
    euclidean_distance_matrix = [[0 for x in range(w)] for y in range(h)] 
    manhattan_distance_matrix = [[0 for x in range(w)] for y in range(h)] 
    cosine_distance_matrix = [[0 for x in range(w)] for y in range(h)]
    minkowski_distance_matrix = [[0 for x in range(w)] for y in range(h)]
    
    
    for i in range(len(x_coord)):
        for j in range(len(x_coord)):
            p1 = np.array([x_coord[i], y_coord[i], z_coord[i]])
            p2 = np.array([x_coord[j], y_coord[j], z_coord[j]])
    
            euclidean_dist = np.sqrt(np.sum((p1-p2)**2, axis=0))
            manhattan_dist = np.sum(np.abs(p1-p2))
            cosine_dist = distance.cosine(p1, p2)
            minkowski_dist = minkowski_distance(p1, p2, 10)
            
            euclidean_distance_matrix[i][j] = euclidean_dist
            manhattan_distance_matrix[i][j] = manhattan_dist
            cosine_distance_matrix[i][j] = cosine_dist
            minkowski_distance_matrix[i][j] = minkowski_dist


    return euclidean_distance_matrix, manhattan_distance_matrix, cosine_distance_matrix, minkowski_distance_matrix
```
